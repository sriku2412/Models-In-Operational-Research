```python
phone_cost = 200
phone_sell = 800
tab_cost = 350
tab_sell = 1000
budget = 1000000

import gurobipy as gp
from gurobipy import GRB

m = gp.Model("Untitled-1")

x = m.addVar(vtype=GRB.INTEGER, name="x", lb=0, ub=3000) # added lb as it should not be negative and ub
y = m.addVar(vtype=GRB.INTEGER, name="y", lb=0)

m.addConstr(phone_cost*x + tab_cost*y <= budget, "c1")
m.setObjective(phone_sell*x + tab_sell*y, GRB.MAXIMIZE)
m.optimize()

print("celphones : %d" % x.x, "tablets : %d" % y.x)
```

```python
g = gp.Model("Untitled-2")

x1 = g.addVar(vtype=GRB.INTEGER, name="x1", lb=0)
x2 = g.addVar(vtype=GRB.INTEGER, name="x2", lb=0)

g.addConstr(4*x1 +6*x2 <= 48, "c1")
g.addConstr(2*x1 + 2*x2 <=18, "c2")
g.addConstr(2*x1 + 1*x2 <= 16, "c3")
g.setObjective(34*x1 + 40*x2, GRB.MAXIMIZE)
g.optimize()

print("x1 : %d" % x1.x, "x2 : %d" % x2.x)
```
